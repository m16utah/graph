```python
"""
предустановки игры:
разерешение 1600*1900 (125%)
нахождение:правый верхний угол
2222,980 / 3840,760
region=(2222,60,3840,980)
лево верх = 2200,50,500,450
лево низ = 2200,500,500,400
сред верх = 2900,330,150,300
сред низ = 2900,600,200,350



"""
import pyautogui    #работа с картинками
import time         #таймеры
import win32api     #для управления мышью, pyautohui не всегда срабатывает клик
import win32con     #для управления мышью, pyautohui не всегда срабатывает клик
import win32com.client     #для управления нажатием HOME
import pyttsx3      #звук
import os           #переход в спящий режим
import pytesseract as tess  #расознование текста
tess.pytesseract.tesseract_cmd  = r'C:\Program Files\Tesseract-OCR\tesseract.exe' #расознование текста
#tess.pytesseract.tesseract_cmd  = r'C:\Program Files\Tesseract-OCR\4.0-with-LSTM\tesseract.exe'
from PIL import Image #расознование текста
import re          #парсинг цен
import pyperclip   #работа с буфером
import random      #рандом

""" Change_price
функция изменеия стоимости.
работает при открытом окне "торговая система"
В секторе "Продавцы"(Торговая система) находим цена, которая есть на расстоянии "Станция".
Если строка "Станция" не имеет синего фона - то продолжать, иначе конец функции
переводим в цифру сумму в столбике "Цена" 
в окне "заказы в торговой системе"
выделяем изначальный товара в "продажа"(заказы в торговой система)  - повой мышкой -
изменить ордер 
ввод текста (пока не понятно как, или через буфер обмена, или по цифрам)
"""


def pause_random(x):
    time.sleep(random.uniform(x,x*2))

def click(x,y,t): # просто клик мышки по координатам ЛЕВОЙ
    print("click")
    win32api.SetCursorPos((x,y))
    time.sleep(0.1)
    win32api.mouse_event(win32con.MOUSEEVENTF_LEFTDOWN,x,y,0,0)
    time.sleep(t)
    win32api.mouse_event(win32con.MOUSEEVENTF_LEFTUP,x,y,0,0)
    
def click_R(x,y,t): #  клик мышки по координатам ПРАВОЙ кнопкой
    print("click_R")
    win32api.SetCursorPos((x,y))
    time.sleep(0.1)
    win32api.mouse_event(win32con.MOUSEEVENTF_RIGHTDOWN,x,y,0,0)
    time.sleep(t)
    win32api.mouse_event(win32con.MOUSEEVENTF_RIGHTUP,x,y,0,0)

def LUP(): #выход на 1 позицию
    print("LUP")
    print("поиск продажа")
    #Проверяем наличие вкладки"Торговая система"
    pict = pyautogui.locateOnScreen('prodazha.png', confidence=0.9,region=(2222,60,3840,980)) #без левого столба "заказы в торг сист"
    if pict==None:
        print("не найдена картинка Торговая система")
    else:
        pause_random(0.3)
        x,y=pyautogui.center(pict)
        y=y+100
        click(x,y,0.1)
        pause_random(0.3)
        shell = win32com.client.Dispatch("WScript.Shell")
        shell.SendKeys("{HOME}")
        
def B_LUP(): #выход на 1 позицию
    print("B_LUP")
    print("поиск покупка")
    #Проверяем наличие вкладки"Торговая система"
    pict = pyautogui.locateOnScreen('pokupka.png', confidence=0.9,region=(2200,50,1000,1000)) #без левого столба "заказы в торг сист"
    if pict==None:
        print("не найдена картинка Покупка")
    else:
        pause_random(0.3)
        x,y=pyautogui.center(pict)
        y=y+100
        click(x,y,0.1)
        pause_random(0.3)
        shell = win32com.client.Dispatch("WScript.Shell")
        shell.SendKeys("{HOME}")
        
def menu_prodazha():
    print("menu_prodazha")
    pict = pyautogui.locateOnScreen('i.png', confidence=0.9,region=(2200,50,500,450))
    if pict==None:
        print("не найдена картинка i")
    else:
        pause_random(0.2)
        x,y=pyautogui.center(pict)
        x=x-50
        click(x,y,0.1)
        time.sleep(0.2)
        click(x,y,0.1)
        click(x,y,0.1) #тройной клик по позиции слева
        analiz_cen(x,y)
        
def B_menu_prodazha():
    print("B_menu_prodazha")
    pict = pyautogui.locateOnScreen('i.png', confidence=0.9,region=(2200,500,500,400))
    if pict==None:
        print("не найдена картинка i")
    else:
        pause_random(0.1)
        x,y=pyautogui.center(pict)
        x=x-50
        click(x,y,0.1)
        time.sleep(0.2)
        click(x,y,0.1)
        click(x,y,0.1) #тройной клик по позиции слева
        B_analiz_cen(x,y) #возврат координат левый столб
    
def menu_prodazha_blue():
    print("menu_prodazha_blue")
    pause_random(0.4)
    pict = pyautogui.locateOnScreen('i_blue.png', confidence=0.90,region=(2200,50,500,450))
    if pict==None:
        print("не найдена картинка i_blue")
    else:
        pause_random(0.1)
        x,y=pyautogui.center(pict)
        x=x-50
        y=y+10
        click(x,y,0.1)
        time.sleep(0.2)
        click(x,y,0.1)
        click(x,y,0.1) #тройной клик по позиции слева
        analiz_cen(x,y)
        
def B_menu_prodazha_blue():
    print("menu_prodazha_blue")
    pause_random(0.4)
    pict = pyautogui.locateOnScreen('i_blue.png', confidence=0.90,region=(2200,500,500,400))
    if pict==None:
        print("не найдена картинка i_blue")
    else:
        pause_random(0.1)
        x,y=pyautogui.center(pict)
        x=x-50
        y=y+10
        click(x,y,0.1)
        click(x,y,0.1)
        time.sleep(0.1)
        click(x,y,0.1) #тройной клик по позиции слева
        B_analiz_cen(x,y)
    
def analiz_cen(x,y):
    print("analiz_cen")
    pause_random(0.9)       
    pict = pyautogui.locateOnScreen('Stancia_blue_black.png', confidence=0.99,region=(2850,330,150,300))
    pict2 = pyautogui.locateOnScreen('Stancia_blue.png', confidence=0.99,region=(2850,330,150,300))
    if pict!=None or (pict==None and pict2==None):    #если синяя на 2 месте или вообще нет
        print('меняем цену:')
        pict = pyautogui.locateOnScreen('Stancia_black.png', confidence=0.9,region=(2850,330,150,300)) #сред низ
        if pict==None:
            print("Stancia_black не найдена")
        else:
            x1,y1=pyautogui.center(pict)
            pause_random(0.3)
            click(x1,y1,0.2)
            time.sleep(0.5)
            win32api.keybd_event (win32con.VK_LCONTROL,1)
            time.sleep(0.3)
            shell = win32com.client.Dispatch("WScript.Shell")
            time.sleep(0.3)
            shell.SendKeys("{C}")
            pause_random(0.4)
            shell.SendKeys("^")
            pause_random(0.2)
            shell.SendKeys("^")
            pause_random(0.2)
            text=pyperclip.paste()
            pause_random(0.3)
            print ("цена скопирована")
            text = re.sub(r'Станция	','', text)
            text = text.replace(u'\xa0', u' ')
            text = re.sub(r' ','', text)
            text = re.sub(r'[xa]','', text)
            print (text)
            text = re.sub(r'ISK[^`]{,1000}','', text)
            #конвертация с проверкой
            try:
                a=float(text)
            except ValueError:
                print('Ошибка. скопирован текст')
                a=0
            else:         
                time.sleep(0.1)
                a=a-0.01
                print(a)   # минимальная цена продажи (FLOAT)      - определиться или буфер или переменная!!!
                pyperclip.copy (a)
                change_price(x,y)
    else:
        print("Изменение цен не требуется")
        click(x,y,0.1)
        
def B_analiz_cen(x,y):
    print("B_analiz_cen")
    pause_random(0.9)       
    pict = pyautogui.locateOnScreen('Stancia_green_black.png', confidence=0.99,region=(2850,600,200,350))
    pict2 = pyautogui.locateOnScreen('Stancia_blue.png', confidence=0.99,region=(2850,600,200,350))
    if pict!=None or (pict==None and pict2==None):    #если синяя на 2 месте или вообще нет
        print('меняем цену:')
        pict = pyautogui.locateOnScreen('Stancia_blue.png', confidence=0.9,region=(2850,600,200,350)) #сред низ
        if pict==None:
            print("Stancia_blue не найдена")
        else:
            x1,y1=pyautogui.center(pict)
            pause_random(0.3)
            click(x1,y1,0.2)
            time.sleep(0.5)
            win32api.keybd_event (win32con.VK_LCONTROL,1)
            time.sleep(0.3)
            shell = win32com.client.Dispatch("WScript.Shell")
            time.sleep(0.3)
            shell.SendKeys("{C}")
            pause_random(0.4)
            shell.SendKeys("^")
            pause_random(0.2)
            shell.SendKeys("^")
            pause_random(0.2)
            text=pyperclip.paste()
            pause_random(0.3)
            print ("цена скопирована")
            text = re.sub(r'Станция	','', text)
            text = text.replace(u'\xa0', u' ')
            text = re.sub(r' ','', text)
            text = re.sub(r'[xa]','', text)
            print (text)
            text = re.sub(r'ISK[^`]{,1000}','', text)
            #конвертация с проверкой
            try:
                a=float(text)
            except ValueError:
                print('Ошибка. скопирован текст')
                a=0
            else:         
                time.sleep(0.1)
                a=a+0.011
                print(a)   # минимальная цена продажи (FLOAT)      - определиться или буфер или переменная!!!
                pyperclip.copy (a)
                B_change_price(x,y)
    else:
        print("Изменение цен не требуется")
        click(x,y,0.1)

def change_price(x,y):
    print("change_price")
    pause_random(0.2)
    click_R(x,y,0.2)
    time.sleep(0.4)
    pict = pyautogui.locateOnScreen('izmenit_order.png', confidence=0.95,region=(2200,50,800,600)) #без левого
    if pict==None:
        print("ОШИБКА! не найдена картинка измененить ордер")
    else:
        pause_random(0.1)
        click (x+20,y+15,0.2) #в подменю "изменить цену"
        #прожимаем Del
        win32api.keybd_event (win32con.VK_LCONTROL,1)
        pause_random(0.3)
        shell = win32com.client.Dispatch("WScript.Shell")
        shell.SendKeys("{V}")
        pause_random(0.3)
        shell.SendKeys("^")
        pause_random(0.3)
        shell.SendKeys("{ENTER}")
        pause_random(0.3)
        pict = pyautogui.locateOnScreen('da.png', confidence=0.9,region=(2222,60,1000,1000))
        if pict:
            x1,y1=pyautogui.center(pict)
            time.sleep(0.5)
            click(x1,y1,0.2)
        else:
            print("не найдена ДА")
            
def B_change_price(x,y):
    print("B_hange_price")
    pause_random(0.2)
    click_R(x,y,0.2)
    time.sleep(0.4)
    pict = pyautogui.locateOnScreen('izmenit_order.png', confidence=0.95,region=(2200,500,700,400)) #без левого
    if pict==None:
        print("ОШИБКА! не найдена картинка измененить ордер")   
    else:
        pause_random(0.1)
        x1,y1=pyautogui.center(pict)
        click (x1,y1,0.2) #в подменю "изменить цену"
        #прожимаем Del
        win32api.keybd_event (win32con.VK_LCONTROL,1)
        pause_random(0.3)
        shell = win32com.client.Dispatch("WScript.Shell")
        shell.SendKeys("{V}")
        pause_random(0.3)
        shell.SendKeys("^")
        pause_random(0.3)
        shell.SendKeys("{ENTER}")
        pause_random(0.3)
        pict = pyautogui.locateOnScreen('da.png', confidence=0.9,region=(2222,60,1000,1000))
        if pict:
            x1,y1=pyautogui.center(pict)
            time.sleep(0.5)
            click(x1,y1,0.2)
        else:
            print("не найдена ДА")

def run():
    print("Start_SELL >>>>>>>>>>>>>>>>>>>")
    i=0
    pause_random(0.5)
    pict = pyautogui.locateOnScreen('lup_end.png', confidence=0.9,region=(2200,50,500,500))
    LUP()
    while pict==None: 
        pause_random(0.1)
        print('Новый цикл ====')
        if i==0:
            menu_prodazha()
        else:
            menu_prodazha_blue()
        i=+1
        pict = pyautogui.locateOnScreen('lup_end.png', confidence=0.9,region=(2200,50,500,500))
        time.sleep(0.1)
        shell = win32com.client.Dispatch("WScript.Shell")
        shell.SendKeys("{DOWN}")
    pause_random(0.2)    
    shell = win32com.client.Dispatch("WScript.Shell")
    shell.SendKeys("{HOME}")

def B_run():
    print("Start_BUY >>>>>>>>>>>>>>>>>>>")
    i=0
    pause_random(0.5)
    pict = pyautogui.locateOnScreen('lup_end.png', confidence=0.9,region=(2200,600,500,400)) #лев низ
    B_LUP()
    while pict==None: 
        print('Новый цикл ====')
        pause_random(0.1)
        if i==0:
            B_menu_prodazha()
        else:
            B_menu_prodazha_blue()
        i=+1
        pict = pyautogui.locateOnScreen('lup_end.png', confidence=0.9,region=(2200,600,500,400)) #лев низ
        time.sleep(0.1)
        shell = win32com.client.Dispatch("WScript.Shell")
        shell.SendKeys("{DOWN}")
    pause_random(0.2)    
    shell = win32com.client.Dispatch("WScript.Shell")
    shell.SendKeys("{HOME}")

time.sleep(5)
i=0    
for i in range(0,100):   
    print ("новый цикл ================================= ",i)
    run()
    B_run()
    pause_random(10)
print("END")
```
